```python
def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s
```

```python
def colorize_displacement(img):

    import matplotlib.colors
    import math
    disp = img.copy()
    disp[...,0:2] = disp[...,0:2]*disp[...,2][...,None]
    mask = disp[...,3]<0.5
    
    a = (np.arctan2(disp[...,0], disp[...,1]) / math.pi + 1) / 2
    r = np.sqrt(disp[...,0] ** 2 + disp[...,1] ** 2)
    s = r / np.max(r)
    hsv_color = np.stack((a, s, np.ones_like(a)), axis=-1)
    rgb_color = matplotlib.colors.hsv_to_rgb(hsv_color)
    rgb_color[mask] = 1
    return rgb_color
```

```python
def _get_instance(module, config, *args):
    """
    Args:
        module (module): The python module.
        config (Box): The config to create the class object.

    Returns:
        instance (object): The class object defined in the module.
    """
    cls = getattr(module, config.name)
    kwargs = config.get('kwargs')
    return cls(*args, **config.kwargs) if kwargs else cls(*args)
```
